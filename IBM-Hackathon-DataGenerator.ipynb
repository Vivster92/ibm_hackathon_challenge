Load the data uploaded to the current Colab instance under /content/sample_data/

```
with open("/content/sample_data/NonZoonoticVirusesStringent0_0_GlobalViromeProject.fasta") as f:
    NonZoo_raw_data  = f.read()
    
with open("/content/sample_data/ZoonoticVirusSequencesStringentGlobalViromeProject.fasta") as g:
    Zoo_raw_data  = g.read()
```

Preprocessing the data

```
NonZoodata = NonZoo_raw_data.split(">")
Zoodata = Zoo_raw_data.split(">")

wrong_stringNonZoo = "NC_001876.1 Bovine adenovirus 3 complete genome\n"
wrong_stringZoo = "NC_003977.2 Hepatitis B virus (strain ayw) genome\n"
correct_stringNonZoo = "NC_001876.1 Bovine adenovirus 3, complete genome\n"
correct_stringZoo = "NC_003977.2 Hepatitis B virus (strain ayw), genome\n"

NonZoodata[58] = NonZoodata[58].replace(wrong_stringNonZoo,correct_stringNonZoo)
Zoodata[68] = Zoodata[68].replace(wrong_stringZoo,correct_stringZoo)
# dump the epmpy string in position [0]
NonZoodata2 = NonZoodata[1:]
Zoodata2 = Zoodata[1:]
```

**Generate train and test data and save in separate files.**

```
zoo_train = Zoodata2[0:70]
nonzoo_train = NonZoodata2[0:70]
zootrain_data = ">"
zootrain_data = zootrain_data.join(zoo_train)
zootrain_data = ">" + zootrain_data
nonzootrain_data = ">"
nonzootrain_data = nonzootrain_data.join(nonzoo_train)
nonzootrain_data = ">" + nonzootrain_data
```

```
with open("/content/sample_data/NonZoonoticVirusesTrain.fasta","w+") as l:
    l.write(nonzootrain_data)
```

```
with open("/content/sample_data/ZoonoticVirusesTrain.fasta","w+") as m:
    m.write(zootrain_data)
```

```
zoo_test = Zoodata2[70:]
nonzoo_test = NonZoodata2[70:]
zoo_test_label = [0]*len(zoo_test) #Zootonic viruses as Class 0.
nonzoo_test_label = [1]*len(nonzoo_test) #NonZootonic viruses as Class 1.
test_x = zoo_test + nonzoo_test
test_y = zoo_test_label + nonzoo_test_label
```

```
import random
random.seed(1)
idx = [x for x in range(0,len(test_y))]
random.shuffle(idx)
x_test = [test_x[x] for x in idx]
y_test = [str(test_y[x]) for x in idx]
```

```
X = ">"
X = X.join(x_test)
X = ">" + X
y = ","
y = y.join(y_test)

with open("/content/sample_data/VirusesTestInput.fasta","w+") as n:
    n.write(X)
    
with open("/content/sample_data/VirusesTestLabel.fasta","w+") as o:
    o.write(y)
```
