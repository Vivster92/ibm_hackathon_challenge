# **Load the predicted and ground-truth classes**

```
with open("/content/sample_data/VirusesTestPrediction.fasta") as f:
    pred_raw_data  = f.read()
    
with open("/content/sample_data/VirusesTestLabel.fasta") as g:
    ground_raw_data  = g.read()
```

```
predictions = pred_raw_data.split(",")
y_pred = [int(x) for x in predictions]
groundtruth = ground_raw_data.split(",")
y_test = [int(x) for x in groundtruth]
```

```
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import pandas as pd

print("Confusion matrix\n")
print(pd.crosstab(pd.Series(y_test, name='Actual'), pd.Series(y_pred, name='Predicted')))
def get_metrics(y_test, y_predicted):
    accuracy = accuracy_score(y_test, y_predicted)
    precision = precision_score(y_test, y_predicted, average='weighted')
    recall = recall_score(y_test, y_predicted, average='weighted')
    f1 = f1_score(y_test, y_predicted, average='weighted')
    return accuracy, precision, recall, f1
accuracy, precision, recall, f1 = get_metrics(y_test, y_pred)
print("accuracy = %.3f \nprecision = %.3f \nrecall = %.3f \nf1 = %.3f" % (accuracy, precision, recall, f1))
```
